```python
#dependencies and imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#read csv and turn into dataframe
cities_path = "Resources/cities.csv"
cities_data = pd.read_csv(cities_path)
cities_pd = pd.DataFrame(cities_data, columns = ["City", "Lat", "Lng", "Max Temp", "Humidity", "Cloudiness", "Wind Speed", "Country", "Date"])
cities_pd.head()
```

```python
#Latitude vs. Temperature Plot
plt.scatter(cities_pd["Lat"], cities_pd["Max Temp"], edgecolors ="black", facecolors="steelblue")
plt.title("City Latitude vs Max Temperature (9/23/20)")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.grid(b=True, which = "major", axis = "both", linestyle = "-", color = "grey")

plt.savefig("Output Data/City Lat vs Max Temp.png")
plt.show()
```

```python
#Latitude vs. Humidity Plot
plt.scatter(cities_pd["Lat"], cities_pd["Humidity"], edgecolors = "black", facecolors = "green")
plt.title("City Latitude vs Humidity (9/23/20)")
plt.xlabel("Latitude")
plt.ylabel("Humidity")
plt.grid(b=True, which = "major", axis = "both", linestyle = "-", color = "grey")

plt.savefig("Output Data/City Lat vs Humidity.png")
plt.show()
```

```python
#Latitude vs. Cloudiness Plot
plt.scatter(cities_pd["Lat"], cities_pd["Cloudiness"], edgecolors = "black", facecolors = "skyblue")
plt.title("Citiy Latitude vs Cloudiness (9/23/20)")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness")
plt.grid(b=True, which = "major", axis = "both", linestyle = "-", color = "grey")

plt.savefig("Output Data/City Lat vs Cloudiness.png")
plt.show()
```

```python
#Latitude vs. Wind Speed Plot
plt.scatter(cities_pd["Lat"], cities_pd["Wind Speed"], edgecolors = "black", facecolors = "purple")
plt.title("City Latitude vs Wind Speed (9/23/20)")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph)")
plt.grid(b=True, which = "major", axis = "both", linestyle = "-", color = "grey")

plt.savefig("Output Data/City Latitude vs Wind Speed.png")
plt.show()
```
